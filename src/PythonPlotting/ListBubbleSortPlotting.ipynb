```python
import matplotlib.pyplot as plt
import pandas as pd

filepath = 'Tables/Program.ListBubbleSortComparisonBenchmarking.csv'

df_raw = pd.read_csv(filepath, sep = ';', thousands = ',')

df = df_raw[['Method', 'elemCount', 'Mean [ms]', 'Ratio', 'Allocated [MB]', 'Alloc Ratio']]
```

```python
#Running time


pvt = df.pivot_table(values = 'Mean [ms]', index = 'elemCount', columns = 'Method')
pvt.plot()
plt.xlabel('Element count')
plt.ylabel('Running time mean value, ms')
```
